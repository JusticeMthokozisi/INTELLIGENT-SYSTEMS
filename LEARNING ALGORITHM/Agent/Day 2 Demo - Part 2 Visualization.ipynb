```python
import seaborn as sns
import pandas as pd
import sklearn.metrics as met
```

## Line plot example

```python
sns.set_theme(style="darkgrid")
```

```python
# Load an example dataset with long-form data
fmri = sns.load_dataset("fmri")
```

```python
# Plot the responses for different events and regions
sns.lineplot(x="timepoint", y="signal",
             hue="region", style="event",
             data=fmri)
```

## Regression plot

```python
tips = sns.load_dataset("tips")
```

```python
sns.regplot(x="total_bill", y="tip", data=tips);
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips,
           markers=["o", "x"], palette="Set1");
```

## Back to houses: metrics

```python
#Load up ground truth data
Y = pd.read_csv("truth.csv")
```

```python
#Load up SVR predictions
svr_pred = pd.read_csv("SVR_prediction.csv")
```

```python
#Load up RFR predictions
rfr_pred = pd.read_csv("RFR_prediction.csv")
```

```python
#Display SVR metrics 
display(met.mean_absolute_error(y_true=Y,y_pred=svr_pred))
display(met.r2_score(y_true=Y,y_pred=svr_pred))
err = Y-svr_pred
display(err.hist())
```

```python
#Display RFR metrics
display(met.mean_absolute_error(y_true=Y,y_pred=rfr_pred))
display(met.r2_score(y_true=Y,y_pred=rfr_pred))
err = Y-rfr_pred
display(err.hist())
```
